```python
import pandas as pd
import numpy as np
import folium
#import geopandas as gpd
```

```python
df = pd.read_csv('/content/noise_dus.xlsx - noise_dus.csv', sep = ',')
df
```

```python
df.replace(' ', np.nan, inplace = True)

df.isna()
```

```python
df = df.dropna()

df
```

```python
del df['Index']
```

```python
df['Longitude'] = df['Longitude'].replace(r'\s+', np.nan, regex=True)

df['Longitude'] = df['Longitude'].replace(r'^$', np.nan, regex=True)

df['Longitude'] = df['Longitude'].fillna(-0.99999)

df['Longitude'] = pd.to_numeric(df['Longitude'])
```

```python
df['Latitude'] = df['Latitude'].replace(r'\s+', np.nan, regex=True)

df['Latitude'] = df['Latitude'].replace(r'^$', np.nan, regex=True)

df['Latitude'] = df['Latitude'].fillna(-0.99999)

df['Latitude'] = pd.to_numeric(df['Latitude'])
```

```python
df.info()
```

```python
def label_noise (df):
   if df['DB_TAG'] <= 54:
      return 'Low'
   elif df['DB_TAG'] > 54 and df['DB_TAG'] <= 57:
      return 'Medium'
   elif df['DB_TAG'] > 57 and df['DB_TAG'] <= 61:
      return 'High'
   return 'Very High'
```

```python
df.apply(lambda df: label_noise(df), axis=1)
```

```python
df['tag_noise_label'] = df.apply(lambda df: label_noise(df), axis=1)
```

```python
from branca.element import Figure
```

```python
figMapa = Figure(height = 600, width = 800)
```

```python
map = folium.Map(location = [51.233334, 6.783333], 
                 zoom_start = 8)
```

```python
figMapa.add_child(map)
```

```python
df.tag_noise_label.value_counts()
```

```python
colors = {
    'Low' : 'blue',
    'Very High' : 'red',
    'Medium' : 'green',
    'High': 'orange',
    }
```

```python
feature_group1 = folium.FeatureGroup(name='Low')
feature_group2 = folium.FeatureGroup(name='Very High')
feature_group3 = folium.FeatureGroup(name='Medium')
feature_group4 = folium.FeatureGroup(name='High')
```

```python
for indice, UBS in df.iterrows():

  dB = UBS.DB_TAG
  localizacao = [UBS.Latitude, UBS.Longitude]
  icone = folium.Icon(color = colors[UBS.tag_noise_label])
  
  folium.CircleMarker(location = localizacao, popup = dB, icon = icone, radius=5).add_to(feature_group1)
  folium.CircleMarker(location = localizacao, popup = dB, icon = icone, radius=5).add_to(feature_group2)
  folium.CircleMarker(location = localizacao, popup = dB, icon = icone, radius=5).add_to(feature_group3)
  folium.CircleMarker(location = localizacao, popup = dB, icon = icone, radius=5).add_to(feature_group4)
```

```python
map.add_child(feature_group1)
map.add_child(feature_group2)
map.add_child(feature_group3)
map.add_child(feature_group4)
```
