```python
import pandas as pd

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)

df = pd.read_csv("data/Strassenverkehrslärm.csv")

df.rename({"BEZ;ID;DB_TAG;DB_NACHT;WOHNGEBAEUDE;STRASCHL-HSNR": "raw_column"}, axis = "columns", inplace = True)

#df = df['raw_column'].str.split(';', expand=True)

df[['Bezirk', 'Building ID', 'DB_TAG', 'DB_NACHT','Wohngebaude','HSNR','a','b','c']] = df['raw_column'].str.split(';', expand=True)

df
```

```python
df.dropna()
```

```python
df = df.drop(60692, axis='index')
df
```

```python
df.info(verbose=False, memory_usage='deep')
```

```python
df.drop(['raw_column', 'Wohngebaude', 'HSNR', 'a', 'b', 'c', 'Building ID'], inplace=True, axis=1)
df
```

```python
df.info(verbose=False, memory_usage='deep')
```

```python
def Address_modification(x):
   return x+", Düsseldorf, Germany"
df["Bezirk"] = df["Bezirk"].apply(Address_modification)
```

```python
df
```

```python
df = df.iloc[:20]
df
```

```python
df['Latitude'] = ''
df['Longitude']=''
```

```python
from geopy.geocoders import Nominatim as Nom
geolocator = Nom(user_agent='Tester')

for i in df.index:
    try:
        location = geolocator.geocode(df['Bezirk'][i])
        
        df.loc[i,'Latitude'] = location.latitude
        df.loc[i,'Longitude'] = location.longitude
    except:
        df.loc[i, 'Latitude'] = ' '
        df.loc[i, 'Longitude'] = ' '
```
